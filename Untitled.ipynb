```python
df_clean = df[df["Price"].notna()]
```

```python
df_clean.shape
```

```python
df_clean['borough'].unique()
```

```python
good_boroughs = list(df_clean['borough'].unique())

print(good_boroughs)
```

```python
good_boroughs = good_boroughs[-12:]
```

```python
print(good_boroughs)
```

```python
for good_borough in good_boroughs:
    df_clean = df_clean[df_clean['borough'] != good_borough]
    
```

```python
df_clean.head(20)
```

```python
df_clean['Price'] = df_clean['Price'].astype(float)
```

```python
df_clean['borough'].unique()
```

```python
borough_prices = df_clean[df_clean['borough'] == input()]
borough_prices.plot(kind='line', x='Date', y='Price')
```

```python
df_clean['Year'] = df_clean['Date'].apply(lambda date: date.year)
```

```python
df_clean.head()
```

```python
df_group = df_clean.groupby(by=["borough", "Year"]).mean()
```

```python
df_group.reset_index(inplace=True)
```

```python
df_group.sort_values(by='Price', ascending=False)
```
