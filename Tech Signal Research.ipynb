```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
plt.figure(figsize=(12,4))
sns.heatmap(df.isnull(),cbar=False,cmap='viridis',yticklabels=False)
plt.title('Missing value in the dataset')
```

```python
df.drop(columns={'Share Price Change'}, inplace = True)
```

```python
df.rename(columns={'Avg. SE comp' : 'SE_Comp'}, inplace = True)
```

```python
df.shape
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
df.info()
```

```python
X = df[['GlassdoorR', 'Recommend', 'Employee_growth_twelve', 'Number_employees', 'PE_Ratio']]
y = df[['SE_Comp']] 
X_train, X_test, y_train, y_test=train_test_split(X, y,test_size = 0.30, random_state = 123)
```

```python
df.columns
```

```python
print (X_train.shape, y_train.shape) 
print (X_test.shape, y_test.shape)
```

```python
regression_model = linear_model.LinearRegression() 
```

```python
regression_model= regression_model.fit(X_train, y_train)
```

```python
print('The R-squared is:', regression_model.score(X_train, y_train))
```

```python
regression_model.mse_model
```

```python
regression_model.rsquared_adj
```

```python
print('Coefficients: \n', regression_model.coef_)
```

```python
print('Intercept: \n', regression_model.intercept_)
```

```python
df.columns
```

```python
Train = pd.concat([X_train, y_train], axis=1)
Formula = 'SE_Comp ~ GlassdoorR + Recommend + Employee_growth_twelve + Number_employees + PE_Ratio'
```

```python
regression_model = smf.ols(Formula, 
                           data= Train).fit()
```

```python
print(regression_model.summary())
```

```python
y_pred = regression_model.predict(X_test)
y_pred = pd.DataFrame(y_pred)
```

```python
y_pred
```

```python
y_test = y_test['SE_Comp'].reset_index()
```

```python
y_obs_pred = pd.concat([y_test['SE_Comp'], y_pred], axis =1)

y_obs_pred.columns = ['y_test', 'y_pred']

y_obs_pred['diff'] = y_obs_pred['y_test'] - y_obs_pred['y_pred']

r = sns.scatterplot(x='y_test', y='y_pred', data = y_obs_pred)
```

```python
diff = y_obs_pred['y_pred'] - y_obs_pred['y_test']
diff_sq = diff**2
mean_diff_sq = diff_sq.mean()
RMSE = np.sqrt(mean_diff_sq)
RMSE
```
